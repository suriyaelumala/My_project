```python
q1=df['4s'].quantile(0.25)
q2=df['4s'].quantile(0.75)
IQR=q2-q1
low_bound=q1-(1.5*IQR)
hig_bound=q2+(1.5*IQR)
print(q1,q2)
print(low_bound,hig_bound)
df_cleans=df[(df['4s']>=low_bound)&(df['4s']<=hig_bound)]
```

```python
plt.boxplot(df_cleans['4s'])
plt.show()
```

```python
for i in df_columns_str.columns:
    sb.boxplot(data=df_columns_str,x=i,color='maroon')
    plt.title(i)
    plt.show()
```

```python
df.columns
```

```python
for i in df.columns:
    if i not  in ['matchDate','match_id','match','batsmanName','margin']:
        plt.figure(figsize=(10,5))
        sb.countplot(data=df, x=i,palette='bright')
        plt.title(f'Frequency Distribution of{i}')
        plt.show
```

```python
for i in df.columns:
    if i not in ['batsmanName','match_id','matchDate','match','margin']:
        plt.figure(figsize=(10,6))
        count=df[i].value_counts()
        plt.pie(count,labels=count.index, autopct='%.3f%%',startangle=90)
        plt.title(i)
```
