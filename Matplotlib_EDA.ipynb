### Box plot

#### A box plot  is used to visualize the distribution of a numerical variable and summarize key statistics such as the median, quartiles, and potential outliers. It displays the spread and skewness of the data in a compact form, making it useful for comparing distributions between different categories

```python
df_numerical.describe()
```

```python
Q1=df['6s'].quantile(0.25)
Q3=df['6s'].quantile(0.75)

IQR=Q3-Q1 # Inter_quartile_range

lower_bound=Q1-(1.5*IQR)
upper_bound=Q3+(1.5*IQR)
print(Q1,Q3)
print(lower_bound,upper_bound)
df_clean = df[(df['6s'] >= lower_bound) & (df['6s'] <= upper_bound)]
```

```python
plt.boxplot(df_clean['6s'])
plt.show()
```

#### A count plot displays how many times each category appears in your data. (or) A count plot is a type of bar plot that shows the counts of observations in each categorical bin. It provides a visual representation of the number of occurrences for each category in a dataset

```python
df_1.columns
```

```python
for i in df_1.columns:
    if i not in ['matchDate','match_id','match','batsmanName','margin']:
        plt.figure(figsize=(10, 5)) #(width,height)
        sn.countplot(data=df, x=i, palette='bright')
        plt.title(f'Frequency Distribution of {i}')
        plt.show()
```

### Pie chart

## Line plot

#### Line charts are effective for visualizing trends over time or sequences,
#### showing how data points are connected in a series.
#### They are often used to display stock prices, temperature changes, or any time-series data.

```python
df.groupby('season')['6s'].sum().reset_index()
```

```python
plt.figure(figsize=(10, 5))
average_runs = df.groupby('season')['6s'].sum().reset_index()
print(average_runs)


plt.plot(average_runs['season'].astype('str'),average_runs['6s'],marker='p',linewidth = 1)
plt.title('Total Sixes Over Seasons')
plt.xlabel('Season')
plt.ylabel('Sixes')
```
