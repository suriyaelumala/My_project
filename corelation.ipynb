```python
from sklearn.datasets import load_iris
```

```python
datas2=load_iris()
datas2
```

```python
from sklearn.feature_selection import SelectKBest,f_classif
```

```python
x,y=datas2.data,datas2.target
```

```python
df=pd.DataFrame(datas2.data,columns=datas2.feature_names)
df
```

```python
k_best=SelectKBest(score_func=f_classif,k=2)
k_best
```

```python
x_new=k_best.fit_transform(x,y)
select_ind=k_best.get_support(indices=True)
select_fet=df.columns[select_ind]
select_fet
```

```python
def distinct_fun(*arg):
  new_list=[]
  for i in arg:
    if i not in new_list:
      new_list.append(i)
  return new_list
a=[1,2,3,3,3,4,5]
distinct_fun(a)
```
